### Q2.3 (2pts) Create a Person class using `namedtuple`

**Background:** `collections.namedtuple` is an extention of `tuples` and we can use it to create immutable classes. Remembering `tuples`, they are a simple data structure for grouping arbitrary objects that is immmutable, can store related information that is heterogenous. `Tuple`s are faster than `list`s and we can use `tuple`s as keys of a dictionary of multiple types or in sets while `list`a cannot be used. There are a couple drawbacks to tuples; for one, data in a tuple can only be pulled out by accessing it through integer indexes (similar to list). Another is that ensuring the number of fields and/or type of information stored in multiple tuples is a challenge due to its ad-hoc structure. To improve on these limitations, we can use the enhanced `namedtuple()` from the `colections` library!

`colections` library includes high-performance container datatypes beyond the built-in types  `list`,`dict`, and `tuples`. 
1. `namedtuple()` factory function for creating tuple subclasses with named fields 	
2. `deque` list-like container with fast appends and pops on either end 	
3. `Counter` dict subclass for counting hashable objects 
4. `OrderedDict` dict subclass that remembers the order entries were added 	
5. `defaultdict` dict subclass that calls a factory function to supply missing values

````python
# Answer 2.3 
from collections import namedtuple

Student = namedtuple('Student', ['name', 'age', 'pob'])

Stu1 = Student('Mimi', 25, 'USA')

for p in [Stu1]:
    print('Hello, my name is', p.name, 'and I am', p.age,'years old.' ' I was born in', p.pob)
````

### Q2.4 (2pts) Pick one of the datatypes in the collections library, explain how you might use the datatype and demonstrate with an example. 


I picked deque from collections library because Deques support thread-safe, memory efficient appends and pops from either side of the deque with approximately the same O(1) performance in either direction.

````python
# Answer 2.4 
from collections import deque

# Create a deque
Weekday = deque(["Mon","Tue","Wed"])
print (Weekday)

# Append to the right
print("Adding to the right: ")
Weekday.append("Thu")
print (Weekday)

# append to the left
print("Adding to the left: ")
Weekday.appendleft("Sun")
print (Weekday)

# Remove from the right
print("Removing from the right: ")
Weekday.pop()
print (Weekday)

# Remove from the left
print("Removing from the left: ")
Weekday.popleft()
print (Weekday)

# Reverse the dequeue
print("Reversing the deque: ")
Weekday.reverse()
print (Weekday)
````
