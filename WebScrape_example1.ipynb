```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url = 'http://web.mta.info/developers/turnstile.html'
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
soup.findAll('a')
```

```python
one_a_tag = soup.findAll('a')[36]
link = one_a_tag['href']
```

```python
link
```

```python
download_url = 'http://web.mta.info/developers/'+ link
urllib.request.urlretrieve(download_url,'./'+link[link.find('/turnstile_')+1:]) 
time.sleep(1) #pause the code for a sec
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.bbc.com/news/world-europe-48303733"
```

```python
# We use try-except incase the request was unsuccessful because of 
# wrong URL
try:
   page = urlopen(url)
except:
   print("Error opening the URL")
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
soup
```

```python
content = soup.find('div', {"class": "story-body__inner"})
```

```python
content
```

```python
article = ''
for i in content.findAll('p'):
    article = article + ' ' +  i.text
```

```python
with open('scraped_text_BBC1.txt', 'w') as file:
    file.write(article)
```

# CNN

```python
url = "https://www.cnn.com/2019/05/16/us/chicago-missing-pregnant-woman/index.html"
```

```python
# We use try-except incase the request was unsuccessful because of 
# wrong URL
try:
   page = urlopen(url)
except:
   print("Error opening the URL")
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
#body-text > div.l-container
#body-text > div.l-container > div.zn-body__read-all
content = soup.find('div', {"class": "zn-body__read-all"})
```

```python
content
```

```python
article = ''
for i in content.findAll('div'):
    article = article + ' ' +  i.text
```

```python
with open('scraped_text_CNN.txt', 'w') as file:
    file.write(article)
```
